```python
import csv

data = csv.reader(open("./baseball_stats/current_season/nationals2017.csv"), delimiter=",", quotechar='|')
column1= []

for row in data:
    column1.append(row[7])


def factorial(n):
    if n == 1 or n == 0:
        return 1
    else: return n * factorial(n - 1)

def choose(a, b):
    if a - b < 0:
        return 0
    else: return factorial(a)/(factorial(b) * factorial(a - b))

def ProbNHeadsKTosses(n, k):
    return choose(k, n) * (1/2)**k

def ProbLeastNHeadsKTosses(n, k):
    i = n
    result = 0
    while i <= k:
        result = result + ProbNHeadsKTosses(i, k)
        i = i + 1
    return result

def ProbMostNHeadsKTosses(n, k):
    i = n
    result = 0
    while i >= 0:
        result = result + ProbNHeadsKTosses(i, k)
        i = i - 1
    return result

def StringToArray(record):
    i = 0
    scores = []
    while i < len(record):
        if record[i] == 'L' or record[i] == 'L-wo':
            scores = scores + [0]
            i += 1
        elif record[i] == 'W' or record[i] == 'W-wo':
            scores = scores + [1]
            i += 1
        else :
            i += 1
    return scores

def ImpressiveScore(scores):
    i = 0
    ImpScore = 1
    wins = 0
    games = 0
    bestgames = 0
    bestwins = 0
    while i < len(scores):
        if scores[i] == 1:
            wins += 1
            games += 1
        else:
            games += 1
        if ProbLeastNHeadsKTosses(wins, games) < ImpScore:
            ImpScore = ProbLeastNHeadsKTosses(wins, games)
            bestgames = games
            bestwins = wins
            i += 1
        else:
            i += 1
    print(bestwins, "in", bestgames, "Probability:", ImpScore, "z score:", (2 * bestwins - bestgames)/(bestgames**(1/2)), "Impressiveness score:", 1/ImpScore)

def MostImpressiveStreak(scores):
    j = 0
    i = 0
    ImpScore = 1
    wins = 0
    games = 0
    bestgames = 0
    bestwins = 0
    beststart = 0
    bestend = 0
    while j < len(scores):
        i = j
        wins = 0
        games = 0
        while i < len(scores):
            if scores[i] == 1:
                wins += 1
                games += 1
            else:
                games += 1
            if ProbLeastNHeadsKTosses(wins, games) < ImpScore:
                beststart = len(scores) - (j + games - 1)
                bestend = beststart + games - 1
                ImpScore = ProbLeastNHeadsKTosses(wins, games)
                bestgames = games
                bestwins = wins
                i += 1
            else:
                i += 1
        j += 1
    print(bestwins, "in", bestgames, "probability:", ImpScore, "z score:", (2 * bestwins - bestgames)/(bestgames**(1/2)),
          "impressiveness score:", 1 / ImpScore,"Game", beststart, "to", bestend)

    
print(ImpressiveScore(StringToArray(column1)))
print(MostImpressiveStreak(StringToArray(column1)))
```
